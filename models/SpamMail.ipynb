```python
#splitting the data into traingthe data and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

ps = PorterStemmer()

def stemming(content):
    # keep only alphabets
    stemmed_content = re.sub('[^a-zA-Z]', ' ', content)

    # convert to lowercase
    stemmed_content = stemmed_content.lower()

    # split into words
    stemmed_content = stemmed_content.split()

    # remove stopwords & apply stemming
    new_list=[]
    for word in stemmed_content:
        if word not in stopwords.words('English'):
            new_list.append(ps.stem(word))

    # join back to string
    stemmed_content = ' '.join(stemmed_content)

    return stemmed_content
```

```python
#feature extraction for x 
#transform text data into feature vectors that can be used as input to logical reggsirsionmodel
# x_train=x_train.apply(stemming)
# x_test=x_test.apply(stemming)
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)




#convert y_train and y_test values as integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#training the model
model=LogisticRegression()
```

```python
#traing the logistic regression model with traing adata
model.fit(x_train_features,y_train)
```

```python
#evalue accuracy of the the trained model
x_train_prediction=model.predict(x_train_features)
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
print("the accauracy score on traing data is....",x_train_accuracy)
```

```python
x_test_prediction=model.predict(x_test_features)
x_test_accuracy=accuracy_score(x_test_prediction,y_test)
print("the accauracy score on testing data is....",x_test_accuracy)
```
